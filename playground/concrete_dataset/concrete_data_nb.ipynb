# Data Loading 

```python
import pandas as pd
import numpy as np

from sklearn.metrics import mean_squared_error

df = pd.read_csv('concrete_data.csv')
df.head()
```

## A. Build a baseline model 

Use the Keras library to build a neural network with the following:

- One hidden layer of 10 nodes, and a ReLU activation function

- Use the adam optimizer and the mean squared error  as the loss function.

### 1. Split data using `train_test_split` 

Randomly split the data into a training and test sets by holding 30% of the data for testing.

```python
from sklearn.model_selection import train_test_split

def split_data(df):
    
    X = df.drop('Strength', axis=1)
    y = df['Strength']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)

    print(X_train.shape, X_test.shape)
    
    return X_train, X_test, y_train, y_test
```

### 2. Train model on the training data using `50 epochs`

Train the model on the training data using 50 epochs

```python
import tensorflow as tf
```

```python
def regression_model():
    # create model
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(10, activation='relu', 
                                    input_shape=(n_cols,)))
    model.add(tf.keras.layers.Dense(1))

    # compile model
    model.compile(optimizer=tf.keras.optimizers.Adam(), 
                  loss=tf.keras.metrics.mean_squared_error)
    return model
```

### 3. Evaluate the model on the test data

```python
n_cols = X_train.shape[1]
EPOCHS = 50

model = regression_model()
X_train, X_test, y_train, y_test = split_data(df)
model.fit(X_train, y_train, epochs=EPOCHS)
y_pred = model.predict(X_test)
```

```python
print("MSE =", mean_squared_error(y_test, y_pred))
```

Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength.

```python
n_cols = X_train.shape[1]
EPOCHS = 50

mse = []

for i in range(50):
    model = regression_model()
    X_train, X_test, y_train, y_test = split_data(df)
    model.fit(X_train, y_train, epochs=EPOCHS)
    y_pred = model.predict(X_test)
    mse.append(mean_squared_error(y_test, y_pred))
```

```python
print(len(mse))
print('Mean MSE of 50 iterations:', np.mean(mse))
print('Standard Deviation of MSE of 50 iterations:', np.std(mse))
```

# B. Normalize the data 

```python
from sklearn.preprocessing import MinMaxScaler

def normalize(X_train, X_test):

    scaler = MinMaxScaler()

    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled= scaler.transform(X_test)
    del scaler
    
    return X_train_scaled, X_test_scaled
```

```python
X_train_scaled, X_test_scaled = normalize(X_train, X_test)

n_cols = X_train.shape[1]
EPOCHS = 50

model = regression_model()
X_train, X_test, y_train, y_test = split_data(df)
model.fit(X_train_scaled, y_train, epochs=EPOCHS)
y_pred = model.predict(X_test_scaled)
```

```python
print("MSE =", mean_squared_error(y_test, y_pred))
```

```python
n_cols = X_train.shape[1]
EPOCHS = 50

mse = []

for i in range(50):
    model = regression_model()
    X_train, X_test, y_train, y_test = split_data(df)
    
    # normalization
    X_train_scaled, X_test_scaled = normalize(X_train, X_test)
    
    model.fit(X_train_scaled, y_train, epochs=EPOCHS)
    y_pred = model.predict(X_test_scaled)
    mse.append(mean_squared_error(y_test, y_pred))
```

```python
print(len(mse))
print('Mean MSE of 50 iterations:', np.mean(mse))
print('Standard Deviation of MSE of 50 iterations:', np.std(mse))
```

The mean of the MSE has decreased compare to that from Step B.

# D. Increase the number of hidden layers 

- Three hidden layers, each of 10 nodes and ReLU activation function.

```python
def regression_model():
    # create model
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(10, activation='relu', 
                                    input_shape=(n_cols,)))
    model.add(tf.keras.layers.Dense(10, activation='relu'))
    model.add(tf.keras.layers.Dense(10, activation='relu'))
    model.add(tf.keras.layers.Dense(1))

    # compile model
    model.compile(optimizer=tf.keras.optimizers.Adam(), 
                  loss=tf.keras.metrics.mean_squared_error)
    return model
```
